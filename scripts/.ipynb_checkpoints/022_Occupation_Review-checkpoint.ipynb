```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
```

```python
# Load Data
jprdy = pd.read_csv("../data/processed/clean_jeopardy_data.csv")
jprdy.head()
```

```python
jprdy.Occupation = jprdy.Occupation.astype(str)
```

```python
# lower and punctuation 
jprdy['job'] = jprdy['Occupation'].apply(lambda x: " ".join(x.lower() for x in x.split()))
jprdy['job'].head()
#jprdy['job'] = jprdy['job'].str.replace('[^\w\s]','')
```

```python
# stop words
stop = stopwords.words('english')
jprdy['job'] = jprdy['job'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
jprdy['job'].head()
```

```python
# frequent words - maybe edit this manually
freq = pd.Series(' '.join(jprdy['job']).split()).value_counts()[:50]
freq = list(freq.index)
freq
```

```python
job_count = jprdy.groupby('job_category')['Full Name'].agg(['count']).reset_index()
job_count = job_count.sort_values('count')
job_count.to_csv('../Data/job_count.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
```

```python
# attach job category counts

job_scores = jprdy.groupby(['job_category'])['Final Score'].agg(['sum']).reset_index()
job_scores.sort_values(by=['sum'], ascending=False, inplace=True)
job_scores.head(20)
```

```python
avg_job_scores = jprdy.groupby(['job_category'])['Final Score'].agg(['mean']).reset_index()
avg_job_scores.sort_values(by=['mean'], ascending=False, inplace=True)
avg_job_scores.head(20)
```

```python
jprdy['games_played'] = jprdy.groupby('Full Name')['Full Name'].transform('count')

avg_job_scores = jprdy.groupby(['job_category','games_played'])['Final Score'].agg(['mean']).reset_index()
avg_job_scores.sort_values(by=['games_played','mean'], ascending=False, inplace=True)
avg_job_scores.head(15)
```
