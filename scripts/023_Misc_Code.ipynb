```python
# -*- coding: utf-8 -*-
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
```

```python
# Load Data
jprdy = pd.read_csv("../data/processed/clean_jeopardy_data.csv")
jprdy.head()
```

```python
jprdy.Occupation = jprdy.Occupation.astype(str)
```

```python
# lower and punctuation 
jprdy['job'] = jprdy['Occupation'].apply(lambda x: " ".join(x.lower() for x in x.split()))
jprdy['job'].head()
#jprdy['job'] = jprdy['job'].str.replace('[^\w\s]','')
```

```python
# stop words
stop = stopwords.words('english')
jprdy['job'] = jprdy['job'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
jprdy['job'].head()
```
