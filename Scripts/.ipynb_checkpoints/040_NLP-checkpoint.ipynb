```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
```

```python
# Load Data
jprdy = pd.read_csv("../Data/clean_jeopardy_data.csv")
jprdy.head()
```

```python
jprdy.Occupation = jprdy.Occupation.astype(str)
```

```python
# lower and punctuation 
jprdy['job'] = jprdy['Occupation'].apply(lambda x: " ".join(x.lower() for x in x.split()))
jprdy['job'].head()
#jprdy['job'] = jprdy['job'].str.replace('[^\w\s]','')

```

```python
# stop words
stop = stopwords.words('english')
jprdy['job'] = jprdy['job'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
jprdy['job'].head()

```

```python
# frequent words - maybe edit this manually
freq = pd.Series(' '.join(jprdy['job']).split()).value_counts()[:50]
freq = list(freq.index)
freq
```

```python
jprdy['job'].nunique()
```

```python
jprdy['job_category'] = jprdy['job']
```

```python
# update to simple defintions
jprdy.loc[jprdy['job'].str.contains('teacher',regex=False), 'job_category'] = 'teacher'
jprdy.loc[jprdy['job'].str.contains('attorney',regex=False), 'job_category'] = 'attorney'
jprdy.loc[jprdy['job'].str.contains('manager',regex=False), 'job_category'] = 'manager'
jprdy.loc[jprdy['job'].str.contains('writer',regex=False), 'job_category'] = 'writer'
jprdy.loc[jprdy['job'].str.contains('analyst',regex=False), 'job_category'] = 'analyst'
jprdy.loc[jprdy['job'].str.contains('bartender',regex=False), 'job_category'] = 'bartender'
jprdy.loc[jprdy['job'].str.contains('engineer',regex=False), 'job_category'] = 'engineer'

# student categories 
jprdy.loc[jprdy['job'].str.contains('freshman',regex=False), 'job_category'] = 'student'
jprdy.loc[jprdy['job'].str.contains('sophomore',regex=False), 'job_category'] = 'student'
jprdy.loc[jprdy['job'].str.contains('junior',regex=False), 'job_category'] = 'student'
jprdy.loc[jprdy['job'].str.contains('senior',regex=False), 'job_category'] = 'student'

jprdy.loc[jprdy['job'].str.contains('student',regex=False), 'job_category'] = 'student'
# grad students
jprdy.loc[jprdy['job'].str.contains('grad',regex=False), 'job_category'] = 'grad student'

# general jobs
jprdy.loc[jprdy['job'].str.contains('consultant',regex=False), 'job_category'] = 'consultant'
jprdy.loc[jprdy['job'].str.contains('lawyer',regex=False), 'job_category'] = 'lawyer'



```
